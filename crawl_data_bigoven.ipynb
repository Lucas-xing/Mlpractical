```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

```python
def get_pages(number):
    session = requests.Session()
    ua = UserAgent()
    headers = {"User-Agent":ua.random}
    url = "https://www.bigoven.com/recipe/{}".format(number)
    req = session.get(url,headers = headers)
    bsObj_bd = BeautifulSoup(req.text,"html")
    return bsObj_bd
```

```python
def get_soup_data():
    name_ls = []
    content_ls = []
    for i in range(1000):
        print("This is crawling data in page{}".format(i))
        get_soup = get_pages(i)
        name = get_soup.find("title").text
        name_ls.append(name)
        if get_soup.find("textarea",{"class":"form-control"}) is not None: 
            content = get_soup.find("textarea",{"class":"form-control"}).text
            content_ls.append(content)
        else:
            content_ls.append("None")
        if i % 200 ==0:
            print("save....")
            data_df = pd.DataFrame({"titile":name_ls,"content":content_ls})
            data_df.to_csv("recipe_data.csv")
```

```python
# receipe title, ingradients and instructions

name_ls = []
content_ls = []
cooking_time_ls = []
review_ls = [] 
serving_people_ls = []
wt_try_ls = []
calories_ls = []
imageurl_ls = []
for i in range(201565,201570):
    print("This is crawling data in page{}".format(i))
    try:
        get_soup = get_pages(i)
    except:
        print("redirects error")
        continue
    # receipe title,  
    
    name = get_soup.find("title").text
    name_ls.append(name)
#     ingradients and instructions
    if get_soup.find("textarea",{"class":"form-control"}) is not None: 
        content = get_soup.find("textarea",{"class":"form-control"}).text
        content_ls.append(content)
    else:
        content_ls.append("None")
#     review
    if get_soup.find("span",{"class":"rating"}) is not None: 
        review = get_soup.find("span",{"class":"rating"}).text
        review_ls.append(content)
    else:
        review_ls.append("No Reviews")
#     cooking time
    if get_soup.find(attrs={'class':'ready-in rc-opt'}) is not None: 
        ck_time = get_soup.find(attrs={'class':'ready-in rc-opt'}).text.strip()
        cooking_time_ls.append(ck_time)
    else:
        cooking_time_ls.append("None")
#         # the number of serving people
    if get_soup.find(attrs={'class':'yield'}) is not None: 
        serving_people = get_soup.find(attrs={'class':'yield'}).text.strip()
        serving_people_ls.append(serving_people)
    else:
        serving_people_ls.append("None")
# # the number of people want to try
    if get_soup.find(attrs={'class':'trying light-med-grey hidden-xs'}) is not None: 
        wt_try = get_soup.find(attrs={'class':'trying light-med-grey hidden-xs'}).text.strip()
        wt_try_ls.append(wt_try)
    else:
        wt_try_ls.append("None")
# # calories per serving
    if get_soup.find(attrs = {'class':'visible-if-not-pro-member'}) is not None: 
        calories = get_soup.find(attrs = {'class':'visible-if-not-pro-member'}).p.get_text().strip()
        calories_ls.append(calories)
    else:
        calories_ls.append("None")
# image url
    if get_soup.find(attrs = {'class':'recipe-hero-photo img-responsive'}) is not None: 
        imageurl = get_soup.find(attrs = {'class':'recipe-hero-photo img-responsive'}).get("src")
        imageurl_ls.append(imageurl)
    else:
        imageurl_ls.append("None")
```
